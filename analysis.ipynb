```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Administración electrónica

```python
admon_demo = pd.read_excel("admon_demografico.xlsx", index_col=0)
admon_econ = pd.read_excel("admon_socioeconomico.xlsx", index_col=0)
motivos_demo = pd.read_excel("motivos_admon_demografico.xlsx")
motivos_econ = pd.read_excel("motivos_admon_socioeconomico.xlsx")
```

```python
admon_demo
```

```python
por_edad = admon_demo.loc[admon_demo.index.str.strip().str.startswith("Edad:")]
legend = por_edad.columns[1:]
por_edad.columns = ["Total", "Han_usado", "Info_admon", "Descargar_form", "Enviar_form", "No_pudieron"]
por_edad.index = ["16-24", "25-34", "35-44", "45-54", "55-64", "65-74"]
por_edad
```

### Uso de la administración electrónica por edad

```python
plt.figure(figsize=(10,10))
plt.grid()
plt.plot(por_edad.index, por_edad["Han_usado"])
plt.plot(por_edad.index, por_edad["Info_admon"])
plt.plot(por_edad.index, por_edad["Descargar_form"])
plt.plot(por_edad.index, por_edad["Enviar_form"])
plt.legend(legend)
plt.show()
```

### Personas que han necesitado utilizar la administración electrónica y no han podido, por edad

```python
plt.grid()
plt.plot(por_edad.index, por_edad["No_pudieron"])
plt.show()
```
